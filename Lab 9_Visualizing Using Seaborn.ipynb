# Visualizing Renewable Energy Data with Seaborn

### Objectives:
    
1. Pair Plot: Visualizing Relationships Between Renewable Energy Sources and Their Consumption
2. Heatmap: Showing the Correlation Between Different Types of Energy Consumption
3. Box Plot: Comparing the Distribution of Energy Consumption Across Different Sources
4. Violin Plot: Comparing Energy Consumption Distributions
5. Regression Plot: Analyzing the Relationship Between Energy Consumption and Carbon Emissions
6. Facet Grid: Plotting Energy Consumption Over Time for Different Regions
    

### 1. Pair Plot: Visualizing Relationships Between Renewable Energy Sources and Their Consumption

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Creating a DataFrame for pair plot
data = {
    "Solar": [1200, 1500, 1300],
    "Wind": [3400, 3600, 3200],
    "Hydropower": [2900, 3100, 2800],
    "Biomass": [2500, 2700, 2400]
}
df = pd.DataFrame(data)
# Create a pair plot
sns.pairplot(df)
plt.show()
```

The pair plot visualizes relationships between different renewable energy sources, making it easier to spot trends and correlations.

### 2. Heatmap: Showing the Correlation Between Different Types of Energy Consumption

```python
# Create a heatmap to show the correlation matrix
correlation_matrix = df.corr()

# Plot heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Add title
plt.title('Correlation Between Energy Sources')
plt.show()
```

 The heatmap shows the correlation between different energy sources, indicating how closely their consumption is related.

### 3. Box Plot: Comparing the Distribution of Energy Consumption Across Different Sources

```python
# Create a box plot
sns.boxplot(data=df)

# Add title
plt.title('Distribution of Energy Consumption by Source')
plt.show()
```

The box plot helps compare the distribution and detect outliers in the energy consumption data for each source.

### 4. Violin Plot: Comparing Energy Consumption Distributions

```python
# Create a violin plot
sns.violinplot(data=df)

# Add title
plt.title('Violin Plot of Energy Consumption Distributions')
plt.show()
```

The violin plot combines a box plot with a kernel density plot, giving a more detailed view of the distribution shape.

### 5. Regression Plot: Analyzing the Relationship Between Energy Consumption and Carbon Emissions

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data for energy consumption and carbon emissions
energy_values = [100, 200, 300, 400, 500]  # Replace with your actual data
carbon_emissions = [10, 20, 25, 40, 50]   # Replace with your actual data

# Create a DataFrame with energy consumption and carbon emissions
df_reg = pd.DataFrame({
    'Energy Consumption': energy_values,
    'Carbon Emissions': carbon_emissions
})

# Create a regression plot
sns.regplot(x='Energy Consumption', y='Carbon Emissions', data=df_reg)

# Add title
plt.title('Energy Consumption vs Carbon Emissions')
plt.xlabel('Energy Consumption')
plt.ylabel('Carbon Emissions')
plt.show()
```

The regression plot shows the linear relationship between energy consumption and carbon emissions.

### 6. Facet Grid: Plotting Energy Consumption Over Time for Different Regions

```python
# Define the months
months = ['January', 'February', 'March', 'April', 'May', 'June']

# Creating a dataset for different regions and energy consumption over months
df_facet = pd.DataFrame({
    'Month': months * 3,
    'Energy Consumption': [1200, 1300, 1100, 1500, 1400, 1600] * 3,
    'Region': ['North', 'South', 'East'] * 6
})

# Create a FacetGrid plot
g = sns.FacetGrid(df_facet, col='Region', hue='Region', height=4, aspect=1)
g.map(sns.lineplot, 'Month', 'Energy Consumption')

# Add title
plt.subplots_adjust(top=0.85)
g.fig.suptitle('Energy Consumption Over Time by Region')
plt.show()
```

The FacetGrid plot shows energy consumption trends over time for different regions, making it easier to compare multiple plots simultaneously.
